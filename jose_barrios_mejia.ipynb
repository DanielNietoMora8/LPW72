```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Matrix:
    

    def __init__(self, tamano, tipo, numero):
        
        self.tamano = tamano
        self.tipo = tipo
        self.numero = numero
        self.matrices = []

        for i in range(self.numero):
            if self.tipo == "zeros":
                self.matrices.append(np.zeros(self.tamano))
            elif self.tipo == "ones":
                self.matrices.append(np.ones(self.tamano))
            else:
                self.matrices.append(np.random.randint(20, 40, self.tamano))

    def multiply(self):
        
        multiplied_matrices = []

        for i in range(self.numero):
            for j in range(self.numero):
                multiplied_matrices.append(np.dot(self.matrices[i], self.matrices[j]))

        return multiplied_matrices

    def imagen(self):
        
        for matrix in self.matrices:
            plt.imshow(matrix)
            plt.show()

    def sum(self):
       
        sums = []

        for matrix in self.matrices:
            sums.append(np.sum(matrix))

        return sums
```

```python

matrix = Matrix((3, 3), "random", 3)


multiplied_matrices = matrix.multiply()


matrix.imagen()

sums = matrix.sum()


print(sums)
```
